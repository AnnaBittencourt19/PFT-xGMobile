### Listas
- Ordenar: sort
- Adicionar: append (ultima posição) e insert (posição especifica)
- Excluir: pop e remove

```python
minha_lista = [1, "text", 3.14] #criando a lista
print(minha_lista)

minha_lista[0] = 2 #alterando a posição 0 (primeiro elemento)
print(minha_lista)

numeros = [3 , 1, 4]
print("Lista inicial:" , numeros)

numeros.append(2) # Adiciona na ultima posição
numeros.insert(1, 9) # No insert adicionamos em uma posicao especifica no caso o numero 9 foi adicionado na 1 posicao (segundo numero)
print("Lista apos adicionar:" , numeros)

numeros.remove(4)
removido = numeros.pop() #pop remove o ultimo elemento da lista
print("Lista apos remover:" , numeros)
print("Elemento removido com o uso de pop:" , removido)

numeros.sort()
print("Lista ordenada:" , numeros)
numeros.sort(reverse=True) # reverso verdadeiro
print("Lista ordenada ao contrario:" , numeros)
```

### Conjuntos
Funciona do mesmo modo que os conjuntos numericos (faz união (|), interseção (&) e diferença(-)

```python
A = {1,2,3,4}
B = {3,4,5,6}

#Uniao de A e B
uniao = A|B
print(uniao) # desconsidera os numeros repetidos

#Interseção de A e B
intersecao = A&B
print(intersecao) #O que se repete nos dois

#Diferança de A e B
diferenca = A-B
print(diferenca) # 5 e 6 não vão aparecer pq a ordem importa, o A não tem 5 e 6 (se fosse B-A eles apareceriam)
```

#Dicionários
- Tem uma palavra chave e valores
- Estrutura:
  "Palavra_chave" : "valor" , "Palavra_chave2" : "valor"

```python
aluno = {"Nome": "João", "Nota": 8.5, "Curso": "Python"}
print(aluno)

print("Nome: ", aluno['Nome']) #Para imprimir o valor associado a uma palavra chave especifica
print("Keys:", aluno.keys()) #Para imprimir todas as palavras chaves
print("Valores:", aluno.values()) #Para imprimir todos os valores
print("Items: ", aluno.items()) # Imprimir os itens palavra_chave: valor
```

### For
Pega cada item da lista e faz algo com esse item

```python
lista_numeros = [x for x in range(10)] #lista_numeros é uma lista de valores de 0 a 9, a cada valor do range adiciona no x um valor de 0 a 9 (em ordem crescente)
print(lista_numeros)

lista_numeros1 = [x for x in range(10) if x<5]
print(lista_numeros1)
```
