### Função lambda

```python
nome_maiusculo = lambda nome : nome.upper() # uma função que entrega o nome maiusculo
nome = "anna"
print(nome_maiusculo(nome))
```

```python
def calcular_quadrado(x):
  return x**2
print(calcular_quadrado(10))

#nome variavel = lambda variavel que vou usar na função : o que a função deve fazer
quadrado = lambda x: x**2
print(quadrado(10))
```

### Lambda + Pandas

```python
import pandas as pd
df = pd.DataFrame({"texto_bruto":[ "OLÁ MUNDO ", " TEXTO! ", "dados ", " outra frase", "  MAIS UM TEXTO"]})
df['texto_limpo'] = df['texto_bruto'].apply(lambda x: x.lower().strip()) #nova coluna com elementos formatados
print(df)
```

### Recursão

```python
def fatorial(n):
  if n == 1:
    return 1
  return n*fatorial(n-1)

print(fatorial(3))
```

### Alta ordem

```python
resultado = map(lambda x: x+1, [1,2,3,4])
# lambda é só 'lambda x: x+1''
#  [1,2,3,4] é parte do map (a lista que será usada)
print(list(resultado))
# tem que usar o list para converter os elementos resultantes do map em uma lista (o map não faz isso)
```

```python
precos = [100, 250, 400, 50]
com_desconto = list(map(lambda p: p*0.9, precos))
#retorna uma lista com os precos com 10% de desconto
# o map pega uma lista e aplica o lambda em todos os elementos da lista
# p é um elemento da lista (percorre os elementos)
print("Com desconto:", com_desconto)

# filter também precisa usar o list
caros = list(filter(lambda p: p > 150, precos))
# filtra os elementos da lista com valores acima de 150
#mesma coisa que for p in preco: if p > 150:
print("Procutos mais caros:", caros)
```

```python
from functools import reduce

precos = [10,20,30,40,50]
total_estoque = reduce(lambda acumulador, atual: acumulador + atual, precos)
# pega cada elemento e soma todos eles (10+20+30+40+50)
#pega uma lista e faz uma conta com todos os elementos
print("Valor total: ", total_estoque)
```
