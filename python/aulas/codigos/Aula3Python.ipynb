### Controle de Fluxo

```python
for i in range(5):
  if i == 1:
    continue #pula o 1
  if i == 2:
    pass # não acontece nada
  if i == 3:
    break # sai do for
  print(i)
```

### Tratamento de exceções

```python
divisor = 0
try:
  resultado = 10/divisor
except ZeroDivisionError: #só executa se tiver erro (o erro for capturado)
  print("Erro: Divisão por zero!")
else: # só executa se não tiver erro
  print(resultado)
finally: #sempre executa
  print("Fim da operação!")
```

```python
try:
  x = int(input("Numero: "))
except ValueError:
  print("Você digitou algo que não é um número!")
else:
  print(x)
```

```python
conectado = True
try:
  print("Executando tarefa critica")
finally:
  conectado = False
  print("Conexão encerrada:", conectado)
```

### Criar o "próprio" erro

#### Com o uso de classe


```python
class SaldoInsuficienteError(Exception):
    pass
def sacar(valor,saldo):
  if valor > saldo:
    raise SaldoInsuficienteError("Saldo baixo")
  return saldo - valor
sacar(100,50)
```

#### Sem o uso de classe

```python
def sacar(valor,saldo):
  if valor > saldo:
    raise ValueError("Saldo baixo")
  return saldo - valor
sacar(100,50)
```
