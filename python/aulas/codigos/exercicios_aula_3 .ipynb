**Exercício 7**

Crie uma função que recebe um número e retorna seu fatorial.

````python
# Seu código aqui
def fatorial(num):
    if num < 0:
        return "Erro"
    elif num == 0:
        return 1
    else:
        resultado = 1
        for i in range(1, num + 1):
            resultado *= i
        return resultado
print(fatorial(5))
print(fatorial(0))
print(fatorial(-1))
````

**Exercício 18**

Crie uma função chamada `validar_senha(senha)` que deve verificar:

* A senha deve ter pelo menos 8 caracteres
* Deve conter pelo menos um número
* Deve conter pelo menos uma letra maiúscula

Se qualquer uma dessas regras não for cumprida, a função deve lançar uma exceção personalizada chamada:

```
class SenhaInvalidaError(Exception):
    pass
```



A função deve retornar `"Senha válida!"` se tudo estiver correto.

````python
class SenhaInvalidaError(Exception):
    pass

def validar_senha(senha):
    if len(senha) < 8:
        raise SenhaInvalidaError("A senha deve ter pelo menos 8 caracteres")
    numero = False
    for char in senha:
        if char.isdigit():
            numero = True
            break
    if not numero:
        raise SenhaInvalidaError("A senha deve conter pelo menos um número")
    maiuscula = False
    for char in senha:
        if char.isupper():
            maiuscula = True
            break
    if not maiuscula:
        raise SenhaInvalidaError("A senha deve conter pelo menos uma letra maiúscula")
    return "Senha válida!"

try:
  minhaSenha = "Pitty1656465"
  validacao = validar_senha(minhaSenha)
except SenhaInvalidaError as e:
  print({e})
else:
  print(validacao)
finally:
  print("Fim do programa")
````

**Seção 12 — Desafio**

**Exercício 19**

Crie um programa que:

1. Peça nome e idade de 5 pessoas

2. Armazene em uma lista de dicionários

3. Informe quem é a pessoa mais velha

````python
# Seu código aqui
lista_pessoas = []
for i in range(5):
    nome = input("Nome: ")
    while True:
        try:
            idade = int(input("Idade: "))
            if idade >= 0:
                break
        except:
            pass
    lista_pessoas.append({"nome": nome, "idade": idade})
for p in lista_pessoas:
    print(p["nome"], p["idade"])
mais_velha = max(lista_pessoas, key=lambda x: x["idade"])
print("Mais velha:", mais_velha["nome"], mais_velha["idade"])
````
