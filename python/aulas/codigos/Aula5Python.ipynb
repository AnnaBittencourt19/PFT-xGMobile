```python
class RoboBase:
  def __init__(self, nome):
    self.nome = nome
    self.bateria = 100
  def trabalhar(self):
    self.bateria -= 10
    print(self.nome, 'está trabalhando. Bateria:', self.bateria, '%')
meu_robo = RoboBase('Alpha')
meu_robo.trabalhar()
```

```python
class RoboVoador(RoboBase):
  def trabalhar(self):
    print(self.nome, "voando ou caminhando")
meu_robo_voador = RoboVoador('Beta')

def iniciar_tarefa(robo):
  robo.trabalhar()

iniciar_tarefa(meu_robo)
iniciar_tarefa(meu_robo_voador)
```
