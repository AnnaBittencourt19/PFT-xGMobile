Váriaveis sempre devem começar com letra minuscula

Unica coisa de aritmética e operadores que o Python difere do C é o exponenciação que é **

```python
energia_total = 100
consumo_por_hora = 3

#Divisão inteira:
horas_reais = energia_total // consumo_por_hora
print(horas_reais)

#Divisão "normal" (real) com casas decimais:
horas_inteiras = energia_total / consumo_por_hora
print(horas_inteiras)

#Resto divisão:
sobra_energia = energia_total % consumo_por_hora
print(sobra_energia)

#Exponenciação:
area_cobertura = 5 ** 2
print(area_cobertura)
```

**Strings**

De 0 até a (ultima - 1)

texto[inicio : fim : passo]

texto[::-1] inverte a string


```python
texto = "Python"
print(texto[0]) # primeira letra
print(texto[0:2]) #mostra a primeira e a segunda letra
print(texto[::-1])
print(texto[2:]) # da segunda letra em diante
```

**Transformação de textos**

.upper(): Maiusculo

.strip(): Tira espaços

.replace(): Substitui coisas

.split(): Gera uma lista

```python
palavras = 'Tudo posso, mas nem tudo me convém'.split() #.split para verificar as palavras em uma lista
print(palavras)

#W é cada palavra na nossa string palavras
resultado = [[w.upper(), w.lower(), w.strip(), len(w)] for w in palavras]
print(palavras)
for i in resultado:
  print(i)
```

No Python usa ':' e indexação no for para que ele entenda onde começa e onde termina
